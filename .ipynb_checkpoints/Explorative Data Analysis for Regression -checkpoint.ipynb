```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

# Missing Values

```python
features_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]#to get features having NAN values
```

```python
features_with_nan
```

```python
for feature in features_with_nan:
    print('The feature {} has {}% of missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#finding how this features having NAn behaving with SalePrice
for feature in features_with_nan:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)#converting all the NAn values to 1 and others to 0
    data.groupby(feature)['SalePrice'].median().plot.bar()#grouping on feature and median on saleprice
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

# Numerical Variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype !='O']
```

```python
dataset[numerical_features].head()
```

```python
year_features=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_features
```

```python
dataset.groupby('YrSold')['SalePrice'].mean().plot()
plt.xlabel('YrSold')
plt.ylabel('Saleprice')
```

```python
for feature in year_features:
    if feature !='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#discrete variables and continuous variables
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_features+['Id']]
```

```python
print('The discrete features are {}'.format(len(discrete_feature)))
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

```python
#continuous variables
continuous_features=[feature for feature in dataset.columns if len(dataset[feature].unique())>25 and feature not in discrete_feature+year_features+['Id']]
```

```python
print('The continuous features are {}'.format(len(continuous_features)))
```

```python
for feature in continuous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.SalePrice=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.show()  
```

# OUTLIERS

```python
for feature in continuous_features:
    data=dataset.copy()
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel('count')
    plt.show()
```
